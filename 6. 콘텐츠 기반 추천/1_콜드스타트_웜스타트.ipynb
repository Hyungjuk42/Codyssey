## 콜드 스타트(Cold Start)

### 정의
콜드 스타트는 추천 시스템이 새로운 사용자나 아이템에 대해 충분한 정보를 수집하지 못해 적절한 추천을 제공하지 못하는 상황을 의미합니다. 이는 주로 다음과 같은 경우에 발생합니다:


### 사용자 콜드 스타트(User Cold Start)
- **새로운 사용자**: 서비스에 새로 가입한 사용자에 대한 정보가 없을 때. 파악할 수 없는 데이터가 없어 적절한 추천을 제공하기 어려움

#### 해결 방법
- **인지도 기반 추천**: 인기 있는 항목이나 카테고리를 기반으로 추천을 시작 => 일반적인 내용
- **컨텐츠 기반 추천**: 사용자의 행동(영화 시청, 선호 영화 고르기)을 기반으로 비슷한 영화 추천
- **협업 필터링**: 다른 유저들의 행동을 기반으로 추천 -> 비슷한 유저들의 선호 항목 추천
- **하이브리드 추천 시스템**: 협업 필터링과 컨텐츠 기반 필터링을 결합하여 사용. 새로운 사용자나 아이템에 대해서는 컨텐츠 기반 필터링을, 충분한 데이터가 있는 경우에는 협업 필터링을 사용.
- **컨텐츠 자질(Contents Feature)**: 아이템의 특징을 설명하는 정보를 활용하여, 인공신경망을 통해 상호작용 패턴을 예측합니다.

### 아이템 콜드 스타트(Item Cold Start)
- **새로운 아이템**: 시스템에 새로 추가된 아이템에 대한 사용자 인터렉션 데이터가 부족할 때. 유저의 interaction 데이터가 충분하지 않아 추천이 어려움

#### 해결 방법
- **컨텐츠 기반 추천**: 아이템 특성(주연 배우, 장르, 감독...)을 기반으로 추천 제공. 인터렉션 데이터 없이도 가능. 
- **인기도 기반 추천**: 인기 아이템을 추천하여 새로운 아이템에 대한 인지도 높임
- **하이브리드 추천 시스템**: 협업 필터링과 컨텐츠 기반 필터링을 결합하여 사용. 새로운 사용자나 아이템에 대해서는 컨텐츠 기반 필터링을, 충분한 데이터가 있는 경우에는 협업 필터링을 사용.
- **상호작용 자질(Interaction Feature)**: 유사한 웜 아이템의 상호작용 자질을 이용하여 콜드 아이템의 상호작용 자질을 예측합니다.


## 웜 스타트(Warm Start)

### 정의
웜 스타트는 이미 시스템에 존재하는 사용자나 아이템에 대한 데이터가 어느 정도 확보된 상태에서 추천을 제공하는 상황을 의미합니다. 이는 콜드 스타트와 비교하여 다음과 같은 특징을 가집니다:

- **사용자**: 서비스에 가입한 후 어느 정도의 인터렉션 데이터가 쌓인 사용자.
- **아이템**: 시스템에 추가된 후 사용자와의 상호작용이 어느 정도 이루어진 아이템.

### 특징
- **웜 스타트는 콜드 스타트보다 더 많은 오버헤드를 나타냅니다**: 이는 시스템이 이미 실행 중인 앱을 백그라운드에서 포그라운드로 가져와야 하기 때문입니다.
- **상태**: 사용자가 앱을 종료한 후 다시 시작하거나, 시스템이 메모리에서 앱을 제거한 후 다시 시작하는 경우 등이 웜 스타트로 간주됩니다.
