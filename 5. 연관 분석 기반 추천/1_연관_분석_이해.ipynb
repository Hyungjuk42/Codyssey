### **연관 분석**과 **주요 성능 지표**의 이해

#### **정의**

- 데이터베이스에서 항목 간의 관계, 연관성을 발견하는 데이터 마이닝 기법

#### **개념**

- **연관 규칙 학습**
  - 데이터에서 발견된 항목 간의 관계를 If then 규칙으로 표현(If 맥주를 산다 Then 기저귀를 산다)

- **지지도(Support)**
  - 전체 거래에서 특정 항목 또는 항목 집합이 나타나는 비율. 특정 상품 A를 구매할 확률
  - `Support(A) = (A가 포함된 거래 수) / (전체 거래 수)`
  - A에 대한 B의 지지도를 계산할 때에는 아래와 같이 A와 B의 교집합을 구한다.
  - `Support(A -> B) = (A와 B를 모두 포함하는 거래 수) / (전체 거래 수)`

- **최소 지지도(Minimum Support)**
  - 분석에서 고려할 가치가 있는 최소 빈도로 발생하는 항목 집합을 결이하는 기준

- **신뢰도(Confidence)**
  - 항목 A를 구매할 때 B도 구매할 확률
  - `Confidence(A -> B) = (A와 B가 동시에 포함된 거래 수) / (A가 포함된 거래 수)`
  - 1에 가까울 수록 A와 B의 연관성이 높다

- **향상도(Lift)**
  - A와 B가 서로 독립적일 때보다 함께 발생할 확률이 얼마나 높은지. 연관성을 측정
  - 무슨 뜻일까? => A 구매 여부와 상관없이 B를 구매할 확률 대비 A를 구매했을 때 B를 구매할 확률의 증가 비율
  - A 구매 여부와 상관없이 B를 구매할 확률 = Support(B) 
  - A를 구매했을 때 B를 구매할 확률 = Confidence(A->B)
  - `Lift(A -> B) = Confidence(A -> B) / Support(B)`
  - Lift(A -> B) > 1 : 양의 상관관계
  - Lift(A -> B) == 1 : 독립적임
  - Lift(A -> B) < 1 : 음의 상관관계
  
- **독립 확률**
  - 두 사건이 서로에게 영향을 미치지 않을 때의 확률
  - `P(B|A) = P(B)`
  - `P(A ∩ B) = P(A) * P(B)`

#### **알고리즘**

- **Apriori**
  - 어떤 두 아이템 집합이 빈번히 발생하는지 알려주는 일련의 규칙 생성 알고리즘
  
    > Apriori 규칙
    
    - 어떤 항목 집합이 빈발하다면, 그 부분집합도 모두 빈발해야 한다.
  
  - k-항목 집합의 지지도를 계산하고 최소 지지도보다 작은 항목을 제거합니다.
  - k를 1부터 n 까지 늘려가며 반복해 빈발 항목 집합을 계산합니다.


- **FP-Growth**

  - Apriori 알고리즘의 단점을 개선
  - FP-트리 자료구조를 사용해 빈발 항목 집합을 효율적으로 찾는다. 

- **FPV**

  - FP-Growth 알고리즘의 성능 향상판
  - FP-Growth 알고리즘은 수평적 데이터 표현 방식 채택
  - FPV는 수직적 데이터 표현 방식 채택


#### **개념**

- **구매 행렬표**
  - 고객과 상품간 구매관계를 나타내는 행렬

- **FP-트리**
  - 빈발 항목 집합의 효율적 저장/검색을 위한 자료구조

  > 특징
  
  - 경로 공유: 동일한 접두사를 가진 항목 집합은 트리의 경로를 공유
  - Ex) {a, b, c} 와 {a, b, d} 는 a, b 까지 경로 공유
  - 노드 정보: 각 노드는 항목의 이름과 해당 항목까지의 경로가 나타나는 횟수를 저장
- **tid-list**
  - 거래 ID의 목록을 저장

#### **활용 방안**

- 상품 진열, 쿠폰 발행, 추천 시스템 등의 분야에서 사용 가능

#### 연관 규칙 학습의 문제점
- 계산 비용 증가
  - 후보 집합 생성의 과부하: 항목 수가 많아질수록 후보 집합의 수가 기하급수적으로 증가
  - 후보 집합 지지도 계산을 위한 DB 반복 스캔
- 최소 지지도 설정의 어려움
  - 적절한 값 찾기 어려움
  - 드문 항목의 문제
- 해석 및 활용의 어려움
  - 과도한 규칙 생성
  - 인과관계의 부재
