- MovieLens
- KMRD(Korean Movie Rating Dataset)
- Netflix Prize Dataset(Netflix)


### **1. MovieLens**

*   **개요:** GroupLens Research에서 제공하는 가장 널리 사용되는 영화 평점 데이터셋입니다. 다양한 크기의 데이터셋이 제공되며, 연구 및 교육 목적으로 많이 활용됩니다.
*   **특징:**
    *   **다양한 크기:** 10만 개, 100만 개, 1000만 개, 2000만 개 등 다양한 크기의 데이터셋을 제공하여 프로젝트 규모에 맞춰 선택 가능합니다.
    *   **명시적 피드백:** 대부분의 데이터셋은 사용자가 영화에 부여한 평점 (1~5점) 정보를 포함합니다.
    *   **메타데이터 제공:** 영화 제목, 장르, 개봉 연도 등의 메타데이터를 함께 제공합니다.
    *   **널리 사용됨:** 많은 연구와 비교에 사용되어, 알고리즘 성능 비교에 유용합니다.
*   **장점:**
    *   접근성이 뛰어나고 사용하기 편리합니다.
    *   다양한 크기의 데이터셋을 선택할 수 있습니다.
    *   메타데이터가 제공되어 콘텐츠 기반 필터링 연구에도 활용 가능합니다.
*   **단점:**
    *   상대적으로 데이터 크기가 작습니다 (Netflix에 비해).
    *   암묵적 피드백 (예: 시청 시간, 클릭) 정보는 제공하지 않습니다.
    *   데이터가 오래된 경우가 있을 수 있습니다.
*   **적합한 경우:**
    *   추천 시스템 연구 및 교육 목적
    *   알고리즘 성능 비교 및 테스트
    *   소규모 프로젝트 또는 프로토타입 개발





MovieLens small dataset (ml-latest-small)의 모든 인덱스는 다음과 같습니다:

> **movies.csv 파일의 인덱스**
- movieId: 영화의 고유 ID
- title: 영화의 제목 (개봉 연도 포함)
- genres: 영화의 장르 (파이프('|')로 구분된 문자열)

> **ratings.csv 파일의 인덱스**
- userId: 사용자의 고유 ID
- movieId: 영화의 고유 ID
- rating: 사용자가 영화에 매긴 평점 (0.5에서 5.0 사이의 0.5 단위 값)
- timestamp: 사용자가 평점을 매긴 시간 (UTC 기준 UNIX 타임스탬프)

> **tags.csv 파일의 인덱스**
- userId: 사용자의 고유 ID
- movieId: 영화의 고유 ID
- tag: 사용자가 영화에 적용한 태그
- timestamp: 사용자가 태그를 적용한 시간 (UTC 기준 UNIX 타임스탬프)

> **links.csv 파일의 인덱스**
- movieId: 영화의 고유 ID
- imdbId: 영화의 IMDb ID
- tmdbId: 영화의 TMDB ID

이 데이터셋은 9,742개의 영화에 대한 100,836개의 평점과 3,683개의 태그 적용을 포함하고 있으며, 610명의 사용자가 1996년 3월 29일부터 2018년 9월 24일 사이에 생성한 데이터입니다.


### **2. KMRD (Korean Movie Rating Dataset)**

*   **개요:** 한국 영화에 대한 평점 데이터셋입니다. 한국어 기반의 추천 시스템 연구에 유용합니다.
*   **특징:**
    *   **한국 영화 데이터:** 한국 사용자의 한국 영화에 대한 평가를 포함합니다.
    *   **평점 정보:** 사용자가 영화에 부여한 평점 정보를 포함합니다.
    *   **상대적으로 작은 규모:** MovieLens에 비해 데이터 규모가 작습니다.
*   **장점:**
    *   한국어 기반의 추천 시스템 연구에 적합합니다.
    *   한국 영화 시장 분석에 유용합니다.
*   **단점:**
    *   데이터 규모가 상대적으로 작습니다.
    *   영어 기반의 연구에 바로 적용하기 어렵습니다.
    *   메타데이터의 종류가 제한적일 수 있습니다.
*   **적합한 경우:**
    *   한국어 기반 영화 추천 시스템 연구
    *   한국 영화 시장 분석
    *   한국 사용자 대상의 서비스 개발


#### **KMRD small dataset 분석**

KMRD small dataset의 모든 인덱스는 다음과 같습니다:

> **사용자 및 영화 관련 인덱스**
- user: 사용자를 식별하는 고유 인덱스
- movie: 영화를 식별하는 고유 ID

> **평점 관련 인덱스**
- rate: 1부터 10까지의 정수로 표현된 평점
- time: UNIX 시간 형식으로 표현된 평점 부여 시간

> **영화 메타데이터 인덱스**
- title: 영화의 한국어 제목
- title_eng: 영화의 영어 제목
- year: 영화의 개봉 연도
- grade: 영화의 등급 (예: 12세 관람가, 청소년 관람불가 등)

> **영화 추가 정보 인덱스**
- country: 영화의 제작 국가
- genre: 영화의 장르 정보
- casting: 영화의 출연진 정보

> **KMRD small dataset의 주요 통계**
- 사용자 수: 52,028명
- 영화 수: 10,999개
- 고유 영화 수: 600개 (전체의 5.455%)
- 비영점 요소 수: 134,331개
- 희소성(sparsity): 0.9997652606410895

이 인덱스들은 KMRD small dataset의 다양한 파일들에 분산되어 있습니다. 예를 들어, 'movies.txt' 파일에는 영화 메타데이터 인덱스가, 'ratings.csv' 파일에는 사용자, 영화, 평점, 시간 인덱스가, 'countries.csv' 파일에는 영화 ID와 국가 정보가 포함되어 있습니다

### **3. Netflix Prize Dataset (Netflix)**

*   **개요:** 2006년 Netflix에서 주최한 추천 시스템 경연 대회를 위해 공개한 데이터셋입니다. 대규모 데이터와 높은 정확도 목표로 인해 추천 시스템 연구에 큰 영향을 미쳤습니다.
*   **특징:**
    *   **대규모 데이터:** 1억 개 이상의 평점 데이터를 포함합니다.
    *   **명시적 피드백:** 사용자-영화 간의 평점 (1~5점) 정보를 포함합니다.
    *   **익명화된 데이터:** 사용자 및 영화 정보는 익명화되어 제공됩니다.
*   **장점:**
    *   매우 큰 규모의 데이터를 활용하여 현실적인 추천 시스템 연구가 가능합니다.
    *   높은 정확도를 목표로 하는 연구에 적합합니다.
*   **단점:**
    *   메타데이터가 부족합니다. 영화 제목, 장르 등의 정보가 제공되지 않습니다.
    *   암묵적 피드백 정보가 없습니다.
    *   경연 대회 이후 새로운 데이터가 업데이트되지 않았습니다.
*   **적합한 경우:**
    *   대규모 데이터 기반의 추천 시스템 연구
    *   높은 정확도를 목표로 하는 알고리즘 개발
    *   협업 필터링 알고리즘 연구

Netflix Prize dataset small의 모든 인덱스는 다음과 같습니다:

> **사용자 및 영화 관련 인덱스**
- user: 사용자를 식별하는 고유 ID (정수)
- movie: 영화를 식별하는 고유 ID (1부터 17770까지의 정수)

> **평점 관련 인덱스**
- rating: 사용자가 영화에 매긴 평점 (1부터 5까지의 정수)
- date: 평점을 매긴 날짜

이 데이터셋의 주요 특징은 다음과 같습니다:

- 사용자 수: 480,189명
- 영화 수: 17,770개
- 평점 데이터 수: 약 100,480,507개

데이터는 다음과 같은 형식으로 제공됩니다:

`<user, movie, date, rating>`

이 데이터셋은 협업 필터링 알고리즘을 개발하고 테스트하는 데 사용되었으며, 사용자의 이전 평점 데이터를 기반으로 영화에 대한 사용자 평점을 예측하는 것이 주요 목표였습니다[1][2].



**영화 추천 시스템 구축을 위한 적합성 비교**

| 특징             | MovieLens                                 | KMRD                                  | Netflix                                   |
| ---------------- | ---------------------------------------- | ------------------------------------- | ----------------------------------------- |
| 데이터 규모       | 다양함 (작음 ~ 중간)                      | 작음                                  | 매우 큼                                   |
| 피드백 유형       | 명시적 (평점)                           | 명시적 (평점)                          | 명시적 (평점)                             |
| 메타데이터        | 풍부 (제목, 장르, 개봉 연도 등)            | 제한적일 수 있음                        | 부족 (익명화)                            |
| 언어             | 주로 영어                               | 한국어                                 | 언어 정보 없음                            |
| 업데이트 빈도    | 간헐적                                 | 상대적으로 적음                         | 업데이트 중단                             |
| 주요 용도        | 연구, 교육, 알고리즘 비교                | 한국어 기반 연구, 한국 영화 시장 분석 | 대규모 데이터 연구, 높은 정확도 목표     |
| 콘텐츠 기반 연구 | 가능                                     | 제한적일 수 있음                        | 불가능                                     |

**결론**

*   **간단한 테스트, 알고리즘 비교, 교육 목적:** MovieLens가 가장 적합합니다. 다양한 크기의 데이터셋과 풍부한 메타데이터를 제공하기 때문입니다.
*   **한국어 기반 영화 추천 시스템 연구:** KMRD가 유일한 선택지입니다.
*   **대규모 데이터 기반 연구, 높은 정확도 목표:** Netflix 데이터셋이 적합하지만, 메타데이터 부족을 고려해야 합니다.

프로젝트의 목표와 데이터의 특징을 고려하여 적절한 데이터셋을 선택하는 것이 중요합니다. 예를 들어, 콘텐츠 기반 필터링과 협업 필터링을 함께 사용하는 하이브리드 추천 시스템을 구축하고자 한다면, MovieLens가 가장 적합한 선택이 될 것입니다. 한국어 기반의 서비스 개발이라면 KMRD를 고려해야 합니다. 매우 높은 정확도를 요구하는 대규모 시스템이라면 Netflix 데이터를 고려해볼 수 있지만, 메타데이터의 부재를 극복할 방법을 찾아야 합니다.
