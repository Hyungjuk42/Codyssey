## 추천 시스템이란 무엇인가?

> 정의

- 추천 시스템은 사용자 맞춤형 아이템 제공을 목표로 하는 정보 필터링 시스템입니다.
- 핵심 기능: 사용자 선호도 예측
- 기대 가치
    - 과부하 해소
    - 새로운 발견
    - 개인화된 경험을 제공하고
    - 기업은 매출 증대
    - 고객 유지
    - 플랫폼 이용 증대를 달성
    
추천 시스템의 핵심 구성 요소:

- 사용자 (User): 추천의 대상. 사용자 프로필, 과거 행동 (구매, 클릭, 평점 등) 데이터가 활용됩니다.
- 아이템 (Item): 추천의 대상. 상품, 콘텐츠, 뉴스 등 다양한 형태. 아이템의 속성 (장르, 키워드, 가격 등) 데이터가 활용됩니다.
- 데이터 (Data): 사용자-아이템 간의 상호작용 데이터 (평점, 클릭, 구매 등), 사용자/아이템 속성 데이터.
- 추천 알고리즘 (Recommendation Algorithm): 데이터를 분석하여 사용자에게 적합한 아이템을 예측하는 핵심 로직.




## 추천 시스템의 목표 

알고리즘 관점:

- 정확도 (Accuracy): 예측된 추천이 실제 사용자의 선호도와 얼마나 일치하는지.
- 다양성 (Diversity): 사용자에게 너무 유사한 아이템만 추천하지 않고, 다양한 아이템을 추천하여 새로운 발견의 기회를 제공.
- 새로움 (Novelty): 사용자가 이전에 경험하지 못했던 새로운 아이템을 추천.
- 적시성 (Timeliness): 사용자의 최근 관심사나 트렌드를 반영하여 추천.
- 설명 가능성 (Explainability): 추천 이유를 사용자에게 설명하여 신뢰도를 높임.

사용자 관점:

- 정보 과부하 해소: 인터넷에는 너무나 많은 정보가 존재하기 때문에, 사용자가 원하는 정보를 직접 찾기 어렵습니다. 추천 시스템은 사용자가 원하는 정보를 쉽게 찾도록 도와줍니다.
- 새로운 발견의 기회 제공: 사용자가 미처 알지 못했던 새로운 아이템이나 콘텐츠를 발견할 수 있도록 도와줍니다. 이를 통해 사용자의 만족도를 높이고, 경험을 풍부하게 만들어 줍니다.
- 개인화된 경험 제공: 사용자의 취향에 맞는 맞춤형 추천을 통해 더욱 편리하고 만족스러운 서비스를 경험할 수 있도록 합니다.

서비스 제공자 (기업) 관점:

- 매출 증대: 사용자가 상품을 더 쉽게 발견하고 구매하도록 유도하여 매출을 증가시킬 수 있습니다.
- 고객 유지율 향상: 사용자에게 맞춤형 서비스를 제공함으로써 고객 만족도를 높이고, 서비스에 대한 충성도를 높여 고객 유지율을 향상시킬 수 있습니다.
- 플랫폼 이용 증대: 사용자가 플랫폼에서 더 많은 시간을 보내도록 유도하여 플랫폼의 활성도를 높일 수 있습니다.
- 데이터 기반 의사 결정: 사용자 데이터를 분석하여 상품 개발, 마케팅 전략 등에 활용할 수 있는 인사이트를 얻을 수 있습니다.

### 협업 필터링 기법(Collaborative filtering)

#### 개요

사용자의 행동, 활동, 선호도에 대한 정보를 분석, 수집 후 다른 사용자와 비교해 예측.

예시) '부산행'에 5점 평점을 준 2명의 사용자 → 사용자 A가 과거 좋아했던 '반도'를 유저 B에게 추천

- 사용자 A와 사용자 B 모두 같은 아이템에 대해 비슷한 또는 같은 평가를 했다.
- 이때, 사용자 A는 다른 아이템에도 비슷한 호감을 나타냈다.
- 따라서, 사용자 A, B의 성향을 비슷할 것이므로, 다른 아이템을 사용자 B에게 추천한다.


> 장점

- Machine Analyzable Content에 의존하지 않음
- 아이템에 대한 이해 없이도 추천 가능

> 순서

- 사용자-아이템 평가 매트릭스 만들기
- 사용자-사용자 유사성 매트릭스, 코사인 유사도 계산
- 유사한 사용자 탐색
- 후보자 생성
- 후보자 점수화

> 예시

- KNN(K-Nearest Neighbors)
분류할 때, 분류 대상과 가까운 k개를 뽑아 그 중 가장 많은 비율을 차지 한 쪽으로 분류
데이터를 이용해서 가까이 있는 K개의 이웃을 찾고 해당 그룹이 가진 속성을 이용해서 새로운 데이터 포인트를 분류, 예측하는 알고리즘

- 거리 측정 방식
    - 유클리드 거리, 맨해튼 거리, 민코프스키 거리

- Pearson correlation(피어슨 상관 계수)
연속형 두 변수간의 선형관계(관련성)를 구하기 위해 사용된다. -1에서 +1사이의 범위값
    - +1 : 완벽한 양의 상관관계,한 변수가 증가할 때 다른 변수도 비례해 증가
    - 0 : 선형 상관관계 없음
    - -1 : 음의 상관관계. 한 변수가 증가할 때 다른 변수는 정확히 비례해 감소

### 콘텐츠 기반 필터링(Content-based filtering)

아이템에 대한 설명과 사용자 선호에 대한 profile을 기반으로 진행
- 핵심: 아이템의 속성 (예: 영화 장르, 키워드)과 사용자의 과거 선호도를 분석하여 유사한 아이템을 추천.
- 알고리즘 예시: TF-IDF, 코사인 유사도 (Cosine Similarity), 키워드 매칭.
- 장점: 새로운 아이템 추천 가능, 사용자 선호도 변화에 민감.
- 단점: 콘텐츠 정보가 부족하면 성능 저하, 과적합 가능성.

하이브리드 방식 (Hybrid Approach):

- 핵심: 콘텐츠 기반 필터링과 협업 필터링의 장점을 결합하여 단점을 보완.
- 예시: 콘텐츠 기반 필터링으로 초기 추천 제공 후, 협업 필터링으로 개선.
- 장점: 높은 정확도, 콜드 스타트 문제 완화 가능.
- 단점: 복잡성 증가.


딥러닝 기반 추천 (Deep Learning based Recommendation):

- 핵심: 심층 신경망 (DNN), 순환 신경망 (RNN), 컨볼루션 신경망 (CNN) 등을 이용하여 복잡한 사용자-아이템 관계를 모델링.
- 예시: AutoRec, Neural Collaborative Filtering (NCF), Wide & Deep Learning.
- 장점: 높은 성능, 복잡한 패턴 학습 가능.
- 단점: 많은 데이터 필요, 학습 및 디버깅의 어려움, 설명 가능성 부족.정보 필터링 기술의 일종. 특정 사용자가 관심을 가질만한 정보를 추천하는 것이다.
- 주로 쓰이는 기술: 협업 필터링 기법(Collaborative filtering), 콘텐츠 기반 필터링(Content-based filtering)


### 추천 시스템의 주요 알고리즘

- 연관 규칙 (Association Rule)

- 협업 필터링 (Collaborative Filtering)
- 행렬 분해 (Matrix Factorization)
- 하이브리드 방법(Hybrid Methods)
- 딥러닝 기반의 추천 (Deep Learning)
